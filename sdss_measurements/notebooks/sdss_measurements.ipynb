```python
from scipy.stats import norm

mask = (t['SERSIC_N_r']>0.0) & (t['SERSIC_N_r']<5.5) 
n = t['SERSIC_N_r'][mask]

def double_gaussian_like(theta, x=n, x_max=5.5):
    """
    """
    x1, x2, s1, s2, f, = theta
    
    f1 = (1.0 - norm.cdf(x_max, loc=x1, scale=s1))*f
    f2 = (1.0 - norm.cdf(x_max, loc=x2, scale=s2))*(1-f)
    
    p1 = f1*norm.pdf(x, loc=x1, scale=s1)
    p2 = f2*norm.pdf(x, loc=x2, scale=s2)
    p = p1+p2
    L = -1.0*np.sum(np.log(p))
    return L

def double_gaussian_pdf(theta, x=n, x_max=5.5):
    """
    """
    x1, x2, s1, s2, f, = theta
    
    f1 = f
    f2 = (1-f)
    
    p1 = f1*norm.pdf(x, loc=x1, scale=s1)
    p2 = f2*norm.pdf(x, loc=x2, scale=s2)
    p = p1+p2
    return p
```

```python
from scipy.optimize import minimize
x0 = [1.4,3.2,0.3,1.5,0.2]
bounds=[(1,2),(3,4),(0.1,0.5),(1,2),(0.1,1)]
res = minimize(double_gaussian_like, x0, bounds=bounds)
```

```python
theta0 = [1.5, 3.8, 0.3, 0.35, 0.3]
bounds = [(1.0,3.0),(2.0,5.0),(0.1,0.5),(0.1,0.5),(0.2,0.8)]

n = t['SERSIC_N_r']
mask = (n>0) & (n<5.8)
n = np.array(n[mask])

def like(theta, x, max_x):
    
    # extract parameters
    mu1, mu2, sigma1, sigma2, f = theta
    scale1 = mu1
    scale2 = mu2
    s1 = sigma1
    s2 = sigma2
    
    # probability each point belongs to each distribution
    p1 = (f)*lognorm.pdf(x,scale=scale1, s=s1)
    p2 = (1.0-f)*lognorm.pdf(x,scale=scale2, s=s2)
    
    m1 = lognorm.cdf(max_x, scale=scale1, s=s1)
    m2 = lognorm.cdf(max_x, scale=scale2, s=s2)
    p1 = p1/m1
    p2 = p2/m2
    
    # total probability
    p = p1 + p2
    L = -1.0*np.sum(np.log(p))
    
    return L

from scipy.optimize import minimize
res = minimize(like, theta0, bounds=bounds, args=(n,5.8))
```
