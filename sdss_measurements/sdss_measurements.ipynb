## Projected Axis Ratios

```python
shape_key = 'EXP'

from astropy.table import Table
t_1 = Table.read('./data/sample_1_all_'+shape_key+'_shapes.dat', format='ascii')
t_2 = Table.read('./data/sample_2_all_'+shape_key+'_shapes.dat', format='ascii')
t_3 = Table.read('./data/sample_3_all_'+shape_key+'_shapes.dat', format='ascii')
t_4 = Table.read('./data/sample_4_all_'+shape_key+'_shapes.dat', format='ascii')
t_5 = Table.read('./data/sample_5_all_'+shape_key+'_shapes.dat', format='ascii')
t_6 = Table.read('./data/sample_6_all_'+shape_key+'_shapes.dat', format='ascii')

t_1a = Table.read('./data/sample_1_disks_'+shape_key+'_shapes.dat', format='ascii')
t_2a = Table.read('./data/sample_2_disks_'+shape_key+'_shapes.dat', format='ascii')
t_3a = Table.read('./data/sample_3_disks_'+shape_key+'_shapes.dat', format='ascii')
t_4a = Table.read('./data/sample_4_disks_'+shape_key+'_shapes.dat', format='ascii')
t_5a = Table.read('./data/sample_5_disks_'+shape_key+'_shapes.dat', format='ascii')
t_6a = Table.read('./data/sample_6_disks_'+shape_key+'_shapes.dat', format='ascii')

t_1b = Table.read('./data/sample_1_ellipticals_'+shape_key+'_shapes.dat', format='ascii')
t_2b = Table.read('./data/sample_2_ellipticals_'+shape_key+'_shapes.dat', format='ascii')
t_3b = Table.read('./data/sample_3_ellipticals_'+shape_key+'_shapes.dat', format='ascii')
t_4b = Table.read('./data/sample_4_ellipticals_'+shape_key+'_shapes.dat', format='ascii')
t_5b = Table.read('./data/sample_5_ellipticals_'+shape_key+'_shapes.dat', format='ascii')
t_6b = Table.read('./data/sample_6_ellipticals_'+shape_key+'_shapes.dat', format='ascii')

# load disk fraction in each luminosity bin
t = Table.read('./data/disk_fraction.dat', format='ascii')
f_disk_1 = t['f_disk'][0]
f_disk_2 = t['f_disk'][1]
f_disk_3 = t['f_disk'][2]
f_disk_4 = t['f_disk'][3]
f_disk_5 = t['f_disk'][4]
f_disk_6 = t['f_disk'][5]
```

```python
# row and column sharing
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(2, 3, sharey=True, figsize=(7.0,4.0))
plt.subplots_adjust(wspace=0, hspace=0)

#ax1.text(0.2, 3, r'$-17>M_r>-18$')
ax1.set_title(r'$-17>M_r>-18$')
p1 = ax1.errorbar(t_1['q'], t_1['frequency'], t_1['err'], fmt='o', color='black', ms=4)
p2 = ax1.errorbar(t_1a['q'], t_1a['frequency']*f_disk_1, t_1a['err']*f_disk_1, fmt='o', color='blue', ms=4)
p3 = ax1.errorbar(t_1b['q'], t_1b['frequency']*(1-f_disk_1), t_1b['err']*(1-f_disk_1), fmt='o', color='red', ms=4)
ax1.set_xlim([0,1])
ax1.set_ylim([0,3.5])
ax1.set_xticklabels(['','','',''])
ax1.legend([p1,p2,p3],['all','disk','elliptical'], frameon=False, numpoints=1, fontsize=10, loc='upper left')

#ax2.text(0.2, 3, r'$-18>M_r>-19$')
ax2.set_title(r'$-18>M_r>-19$')
ax2.errorbar(t_2['q'], t_2['frequency'], t_2['err'], fmt='o', color='black', ms=4)
ax2.errorbar(t_2a['q'], t_2a['frequency']*f_disk_2, t_2a['err']*f_disk_2, fmt='o', color='blue', ms=4)
ax2.errorbar(t_2b['q'], t_2b['frequency']*(1-f_disk_2), t_2b['err']*(1-f_disk_2), fmt='o', color='red', ms=4)
ax2.set_xlim([0,1])
ax2.set_ylim([0,3.5])
ax2.set_xticklabels(['','','',''])

#ax3.text(0.2, 3, r'$-19>M_r>-20$')
ax3.set_title(r'$-19>M_r>-20$')
ax3.errorbar(t_3['q'], t_3['frequency'], t_3['err'], fmt='o', color='black', ms=4)
ax3.errorbar(t_3a['q'], t_3a['frequency']*f_disk_3, t_3a['err']*f_disk_3, fmt='o', color='blue', ms=4)
ax3.errorbar(t_3b['q'], t_3b['frequency']*(1-f_disk_2), t_3b['err']*(1-f_disk_2), fmt='o', color='red', ms=4)
ax3.set_xlim([0,1])
ax3.set_ylim([0,3.5])
ax3.set_xticklabels(['','','',''])

ax4.text(0.2, 3, r'$-20>M_r>-21$')
ax4.errorbar(t_4['q'], t_4['frequency'], t_4['err'], fmt='o', color='black', ms=4)
ax4.errorbar(t_4a['q'], t_4a['frequency']*f_disk_4, t_4a['err']*f_disk_4, fmt='o', color='blue', ms=4)
ax4.errorbar(t_4b['q'], t_4b['frequency']*(1-f_disk_4), t_4b['err']*(1-f_disk_4), fmt='o', color='red', ms=4)
ax4.set_xlim([0,1])
ax4.set_ylim([0,3.5])
ax4.set_xticks([0,0.25,0.5,0.75])

ax5.text(0.2, 3, r'$-21>M_r>-22$')
ax5.errorbar(t_5['q'], t_5['frequency'], t_5['err'], fmt='o', color='black', ms=4)
ax5.errorbar(t_5a['q'], t_5a['frequency']*f_disk_5, t_5a['err']*f_disk_5, fmt='o', color='blue', ms=4)
ax5.errorbar(t_5b['q'], t_5b['frequency']*(1-f_disk_5), t_5b['err']*(1-f_disk_5), fmt='o', color='red', ms=4)
ax5.set_xlim([0,1])
ax5.set_ylim([0,3.5])
ax5.set_xticks([0,0.25,0.5,0.75])

ax6.text(0.2, 3, r'$-22>M_r>-23$')
ax6.errorbar(t_6['q'], t_6['frequency'], t_6['err'], fmt='o', color='black', ms=4)
ax6.errorbar(t_6a['q'], t_6a['frequency']*f_disk_6, t_6a['err']*f_disk_6, fmt='o', color='blue', ms=4)
ax6.errorbar(t_6b['q'], t_6b['frequency']*(1-f_disk_6), t_6b['err']*(1-f_disk_6), fmt='o', color='red', ms=4)
ax6.set_xlim([0,1])
ax6.set_ylim([0,3.5])
ax6.set_xticks([0,0.25,0.5,0.75,1.0])

plt.show()

fig.savefig('./figures/'+shape_key+'_shape_measurements.pdf', dpi=250)
```

## completeness

```python
# load sample selection
from astropy.table import Table
fpath = '../data/SDSS_Main/'
fname = 'sdss_vagc.hdf5'
t = Table.read(fpath+fname, path='data')

# define cosmology
from default_cosmology import cosmo
#cosmo = FlatLambdaCDM(H0=70, Om0=0.3, Ob0=0.05, Tcmb0=2.7255)

# sample absolute magnitude limit as a function of redshift
from astro_utils.magnitudes import absolute_magnitude_lim
z = np.linspace(0.001,0.5,100)
Mr_lim = absolute_magnitude_lim(z, 17.77, cosmo=cosmo)

# see Yang et al. (2009) eq. 3, 4, and 5
from sdss_utils import maximum_redshift
mm = np.linspace(-24,-16,1000) # sample magnitudes
zz = maximum_redshift(mm) # returns redshift limit

# define (g-r) color
color = np.array(t['ABSMAG_g0.1'] - t['ABSMAG_r0.1'])
```

```python
# color vs magnitude

fig = plt.figure(figsize=(3.3,3.3))
plt.scatter(t['ABSMAG_r0.1'], color,
            s=1, c=color, vmin=0.2, vmax=1.2, cmap='jet')
plt.xlim([-17,-24])
plt.ylim([0,1.5])
plt.xlabel('z')
plt.ylabel(r'$(g-r)_{0.1}$')
plt.colorbar(label=r'$(g-r)_{0.1}$')
plt.show()
```

```python
# magnitude vs redshift

fig = plt.figure(figsize=(3.3,3.3))
plt.scatter(t['Z'], t['ABSMAG_r0.1'],
            s=1, c=color, vmin=0.2, vmax=1.2, cmap='jet')
l1, = plt.plot(z, Mr_lim, '-', color='black')
l2, = plt.plot(zz, mm, '--', color='black')
plt.xlim([0,0.4])
plt.ylim([-17,-24])
plt.xlabel('z')
plt.ylabel(r'$M_{r0.1}$')
plt.colorbar(label=r'$(g-r)_{0.1}$')
plt.legend([l1,l2],['r=17.77 limit','yang + (2009)'],
           loc='lower right', fontsize=10, frameon=False)
plt.show()

fig.savefig('./figures/magnitude_vs_redshift.pdf', dpi=250)
```
